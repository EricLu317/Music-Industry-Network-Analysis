```python
import snap
import numpy as np
import time
import sys

# G_Music_1 = snap.LoadEdgeList(snap.PUNGraph, '/Users/ricky/Desktop/School/FL2018/CSE416/final-project/test_genres.edgelist', 0, 1)
# for node in G_Music_1.Edges():
#     print node.GetId()
G_Music_2 = snap.LoadPajek(snap.PUNGraph, '/Users/ricky/Desktop/School/FL2018/CSE416/final-project/test_genres.net')
# for edge in G_Music_2.Edges():
#     print egde.GetId()


# for NI in G_Music_2.Nodes():
#     print NI.GetId()

print(' * * * * * Feedback * * * * * ')

CmtyV = snap.TCnComV()

t0 = time.clock()

modularity = snap.CommunityGirvanNewman(G_Music_2, CmtyV)
t1 = time.clock()
running_time = t1 - t0

print(' * * * * * Feedback * * * * * ')


for Cmty in CmtyV:
    sys.stdout.write('\n')
    for NI in Cmty:
        sys.stdout.write(str(NI) + ',')


print "The modularity of the network is %f" % modularity
print 'The script ran in %f microseconds' % running_time


with open('betweeness_modularity.txt', 'w') as f:
  f.write('%f' % modularity)

with open('betweeness_timing.txt', 'w') as f:
  f.write('%f' % running_time)


communities = np.array([[node for node in Cmty] for Cmty in CmtyV])
np.savetxt('betweenness_communities.txt', communities)


print(' * * * * * DONE! * * * * * ')
```
