```python
import pickle
G = pickle.load(open("Music_network_w_genre.p", "rb"))
import pandas as pd
import networkx as nx
#Degree Centrality
from networkx import degree_centrality
deg_cent = degree_centrality(G)
df = pd.DataFrame.from_dict(deg_cent, orient='index', columns=['degree_centrality'])
#Betweenness Centrality
from networkx import betweenness_centrality
bet_cent = betweenness_centrality(G, k = 100)
new_df = pd.DataFrame.from_dict(bet_cent, orient='index', columns=['betweenness_centrality'])
df = df.merge(new_df, left_index = True, right_index=True)
#Closeness Centrality
from networkx import closeness_centrality
close_cent = closeness_centrality(G)
new_df = pd.DataFrame.from_dict(close_cent, orient='index', columns=['closeness_centrality'])
df = df.merge(new_df, left_index = True, right_index=True)
```

```python
import pickle
from networkx.classes.function import set_node_attributes
node_to_genre = {}
node_to_artist = {}
G = pickle.load(open("Music_network_w_genre.p", "rb"))

for node in list(G.nodes()):
    node_to_genre[node] = node[1]
set_node_attributes(G, values = node_to_genre, name = "genre")
```

```python
from collections import Counter
nodes = G.nodes()
genre_to_count = Counter([list(x.values())[0] for x in list(dict(nodes.data()).values())])
genre_to_count = {k:v for k,v in genre_to_count.items() if v>10}
sorted_genres = [x[0] for x in sorted(genre_to_count.items(),key = lambda x: x[1], reverse=True)]
filtered_genres_set = set(sorted_genres)
# genre_to_mapping = {x:i for i,x in enumerate(sorted_genres)}
len(sorted_genres)
```

```python
import networkx as nx
import pandas as pd
from networkx.algorithms.assortativity import attribute_mixing_matrix, attribute_mixing_dict
assortativity = nx.attribute_assortativity_coefficient(G,'genre')
# mixing = attribute_mixing_dict(G, attribute = "genre", mapping=genre_to_mapping, normalized=False)
mixing = attribute_mixing_dict(G, attribute = "genre", normalized=False)
normalized_mixing = attribute_mixing_dict(G, attribute = "genre", normalized=True)
genre_cooccurrence = pd.DataFrame.from_dict(mixing).fillna(0)
```

```python
import numpy as np
genre_cooccurrence = pd.DataFrame.from_dict(mixing).fillna(0)
genre_cooccurrence = genre_cooccurrence[sorted_genres].loc[sorted_genres]
genre_to_total = genre_cooccurrence.sum().to_dict()
genre_probability_df = genre_cooccurrence.apply(lambda x: x/[genre_to_total[x] for x in list(x.index)])
in_genre = pd.DataFrame(np.diag(genre_probability_df), index = genre_probability_df.index) #.sort_values(by=0, ascending=False)
in_genre.head(10)
with open("in_genre.csv", 'w') as file:
    file.write(in_genre.head(20).to_csv())
```

```python
different_genre_probability_df = pd.DataFrame.copy(genre_probability_df)
np.fill_diagonal(different_genre_probability_df.values, 0)
different_genre_top_20_df = different_genre_probability_df.stack().nlargest(20)
different_genre_top_20_df.index = pd.DataFrame(list(different_genre_probability_df.stack().nlargest(20).index)).apply(lambda x:x[0]+" -> "+x[1], axis = 1)
print()
with open("top_different_genre.csv", "w") as file:
    file.write(different_genre_top_20_df.to_csv())
```
