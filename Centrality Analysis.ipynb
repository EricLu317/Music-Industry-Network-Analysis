```python
import pickle
with open("centralities_dfs.p", "rb") as file:
    df = pickle.load(file)
```

```python
import pandas as pd
top_20 = pd.DataFrame()
top_20["degree_centrality"] = df.sort_values(by="degree_centrality", ascending=False).degree_centrality.head(20).index
top_20["closeness_centrality"] = df.sort_values(by="closeness_centrality", ascending=False).closeness_centrality.head(20).index
top_20["betweenness_centrality"] = df.sort_values(by="betweenness_centrality", ascending=False).betweenness_centrality.head(20).index
top_20
```

```python
with open("top_20.csv", "w") as file:
    file.write(top_20.to_csv())
```
